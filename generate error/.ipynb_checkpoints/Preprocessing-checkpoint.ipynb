```python
import re

def process_file(input_file, output_file):
    try:
        # 读取输入文件
        with open(input_file, 'r', encoding='utf-8') as f:
            content = f.read()
        
        # 移除多余的空白行
        content = re.sub(r'\n\s*\n', '\n', content)
        
        # 分段处理
        sentences = []
        current_sentence = []
        
        for line in content.split('\n'):
            line = line.strip()
            if line:
                # 跳过XML标签
                if line.startswith('<') and line.endswith('>'):
                    continue
                    
                # 使用正则表达式分割句子
                line_sentences = re.split(r'([。！？!?])', line)
                
                for i in range(len(line_sentences)):
                    if i % 2 == 0:
                        current_sentence.append(line_sentences[i])
                    else:
                        current_sentence.append(line_sentences[i])
                        full_sentence = ''.join(current_sentence).strip()
                        if full_sentence:
                            sentences.append(f"<p>{full_sentence}</p>")
                        current_sentence = []
                
                # 处理没有结束标点的句子
                if current_sentence:
                    full_sentence = ''.join(current_sentence).strip()
                    if full_sentence:
                        sentences.append(f"<p>{full_sentence}</p>")
                    current_sentence = []

        # 写入输出文件
        with open(output_file, 'w', encoding='utf-8') as f:
            f.write('\n'.join(sentences))
            
        print(f"处理完成！输出文件: {output_file}")
        
    except Exception as e:
        print(f"处理文件时出错: {str(e)}")

def main():
    input_file = "output.txt"    # 输入文件路径
    output_file = "output.txt"  # 输出文件路径
    
    process_file(input_file, output_file)

if __name__ == "__main__":
    main()
```

```python
import re

def remove_p_tags(input_file, output_file):
    try:
        # 读取输入文件
        with open(input_file, 'r', encoding='utf-8') as f:
            content = f.read()
        
        # 删除<p>和</p>标签
        content = content.replace('<p>', '')
        content = content.replace('</p>', '')
        
        # 写入输出文件
        with open(output_file, 'w', encoding='utf-8') as f:
            f.write(content)
            
        print(f"处理完成！输出文件: {output_file}")
        
    except Exception as e:
        print(f"处理文件时出错: {str(e)}")

def main():
    input_file = "output.txt"    # 输入文件路径
    output_file = "output.txt"  # 输出文件路径
    
    remove_p_tags(input_file, output_file)

if __name__ == "__main__":
    main()
```

```python
def process_text(text):
    # 将文本按行分割
    lines = text.split('\n')
    
    processed_pairs = []
    current_error = None
    current_correct = None
    
    for line in lines:
        line = line.strip()
        
        # 检查新的错误文本开始
        if line.startswith('错误文本：'):
            # 如果已经有一对完整的错误和正确文本，保存它们
            if current_error is not None and current_correct is not None:
                processed_pairs.append((current_error, current_correct))
            # 开始新的错误文本
            current_error = line[5:].strip()
            current_correct = None
        
        # 检查新的正确文本开始
        elif line.startswith('正确文本：'):
            if current_error is not None:  # 确保有对应的错误文本
                current_correct = line[5:].strip()
                if current_error != current_correct:  # 只保存真正有区别的文本对
                    processed_pairs.append((current_error, current_correct))
                current_error = None
                current_correct = None
    
    # 处理最后一对（如果存在）
    if current_error is not None and current_correct is not None:
        processed_pairs.append((current_error, current_correct))
    
    # 生成输出文本
    output_lines = []
    for error, correct in processed_pairs:
        output_lines.append(f'错误文本：{error}')
        output_lines.append(f'正确文本：{correct}')
        output_lines.append('')  # 添加空行
    
    return '\n'.join(output_lines)

def main():
    # 读取输入文本
    with open('output.txt', 'r', encoding='utf-8') as file:
        text = file.read()
    
    # 处理文本
    result = process_text(text)
    
    # 将结果写入新文件
    with open('output.txt', 'w', encoding='utf-8') as file:
        file.write(result)
    
    # 打印处理结果的统计信息
    pair_count = len(result.split('\n\n')) - 1
    print(f'成功处理了 {pair_count} 对错误/正确文本对')

if __name__ == "__main__":
    main()
```
